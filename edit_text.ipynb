```python
import os

words = []
with open("./Vue/text_eng.txt", mode="r+") as file:
    for line in file:
        cleared_line = []
        for word in line.split(" "):
            word = word.strip()
            word.replace("–", "-")
            word.replace("—", "-")
            if word:
                cleared_line.append(word.strip())
        # print(cleared_line)
        words += cleared_line

words_str = " ".join(words)

with open("./vue/text_eng_cleared.txt", mode="w") as file:
    file.write(words_str)
```
